# 概率建模

```python
def laplace_smooth(counter, c=1):
    N = sum(counter.values())
    Nplus = N + c * (len(counter) + 1)
    
    return lambda word: (counter[word] + c ) / Nplus
```

```python
laplace_prob=laplace_smooth(words_count)
```

# Good-Turing 对于词库中未出现的词，用词库中对应长度的词的概率*prior去估计，若无对应长度，则加上一个bias去调整

```python
# 生成器（训练器）返回的是一个函数
def good_turing_simple_version(counter, bias=1/26, prior=1e-8):
    N=sum(counter.values())
    single_words=map(len, [w for w in counter if counter[w]==1])
    ones=Counter(single_words)
    longest=max(ones)
    def _get_prob(w):
        if w in counter:
            prob=counter[w]/N
        elif len(w) in ones:
            prob=prior*ones[len(w)]/N
        else :
            prob=prior*ones[longest]/N*bias**(len(w)-longest)
        return prob
    return _get_prob
```

```python
gt_prob = good_turing_simple_version(words_count)
```

```python
gt_prob("狼狼")
```

# one-gram model 
cut之后对每个词计算概率并相乘

```python
def language_model_one_gram (string):
    word=cut(string)
    prob_map = list(map(gt_prob, word))
    prob = reduce(lambda x1, x2: x1*x2 ,prob_map)  
    #此类lambda函数要加reduce
    return prob
```

```python
test="""
我喜欢吃酸菜
我喜欢吃狗屎
那里有条河
那里油条河
""".split()
for i in test:
    print("\"{}\"的概率为{}".format(i,language_model_one_gram(i)))
```

```python
need_compared = [
    "今天晚上请你吃大餐，我们一起吃日料 明天晚上请你吃大餐，我们一起吃苹果",
    "真事一只好看的小猫 真是一只好看的小猫",
    "我去吃火锅，今晚 今晚我去吃火锅"
]
for s in need_compared:
    s1, s2 = s.split()
    p1, p2 = language_model_one_gram(s1), language_model_one_gram(s2)
    
    better = s1 if p1 > p2 else s2
    
    print("{} with probability{}\n".format(s1,p1)+
          "{} with probability{}\n".format(s2,p2)+
          "{} is better ".format(better))
```

#  Two gram model

```python
all_two_gram_words = ["".join(valida_tokens[i:i+2]) for i in range(len(valida_tokens[:-2]))]
```

```python
all_two_gram_words[:10]
```

```python
words_count_2_gram = Counter(all_two_gram_words)
```

```python
_2_sum = sum(words_count_2_gram.values())
_2_sum
```

```python
def language_2_gram_generator (counter1, counter2):
    N1 = sum (counter1.values())
    N2 = sum (counter2.values())
    def prob_2_gram_combine (w1, w2):
        if w1+w2 in words_count_2_gram:
            prob = words_count_2_gram[w1+w2]/N1
        else: prob = 1/N1
        return prob
    def get_prob(w1):
        if w1 in counter2:
            prob = counter2[w1]/N2
        else: prob = 1/N2
        return prob
    def get_2_gram_prob(w1,w2):
         return prob_2_gram_combine(w1,w2)/get_prob(w1)
    return get_2_gram_prob
```

```python
gt_2_gram = language_2_gram_generator(words_count_2_gram, words_count)
```

```python
# 当处理首字符时，在内部函数很不方便，不得不外置，期待更好的解决办法
def get_prob(w1):
    N = sum (words_count.values())
    if w1 in words_count:
        prob = words_count[w1]/N
    else: prob = 1/N
    return prob
```

```python
def _2_gramm_district(string):
    words = cut(string)
    prob1 = [get_prob(words[0])]
    prob2 = [gt_2_gram(words[i], words[i+1]) for i in range(1,len(words[:-2]))]
    prob = prob1+prob2
    return reduce(lambda a1, a2: a1*a2 , prob)
```
